```python
import matplotlib.pyplot as plt
import numpy as np
from IPython.display import HTML, Image
from scipy.special import airy
%matplotlib inline
```

```python
def f(x):
    return np.sin(x)
x = np.linspace(0,10)
plt.plot(x, f(x), 'mo')
plt.xlabel('xaxis')
plt.ylabel('yaxis')
plt.title('stuff')
plt.show()
```

```python
x = np.linspace(-np.pi, np.pi)
y = np.sin(x)

plt.figure(figsize=(12,3)) #width hiehgt
plt.subplot(2,3,1) #creates first plot in 2 by 3 grid of plots
plt.plot(x,y)
plt.subplot(2,3,2)
plt.plot(x,y,'r--')
plt.subplot(2,3,3)
plt.plot(x,y,'b--')
plt.subplot(2,3,4)
plt.plot(x,y,'k--')
plt.show()
```

```python
x = np.linspace(0,20)
y = np.sqrt(x)

plt.figure(figsize=(10,5))
plt.subplot(1,3,1)
plt.plot(x,y,'k--', lw = 30)
plt.subplot(1,3,2)
plt.plot(x,y,'g--', linewidth = 2)
plt.subplot(1,3,3)
plt.plot(x,y)
plt.show()
```

```python
x = np.linspace(0,20)
y = np.cos(x)
z = np.cos(x)**2

plt.figure(figsize = (12,3))
plt.subplot(1,3,1)
plt.plot(x,y)
plt.plot(x,z,'k--')
plt.subplot(1,3,2)
plt.plot(z,x)
plt.plot(y,x, 'm--')
plt.subplot(1,3,3)
plt.plot(z,y, 'ro')
plt.plot(y,z)
```

```python
x = np.linspace(0,20)
y = np.cos(x)
z = np.cos(x)**2
a = np.sin(x)

plt.figure(figsize = (12,3))
plt.subplot(1,2,1)
plt.plot(x,y, 'g')
plt.plot(x,z, 'b')
plt.plot(x,a, 'r')

plt.subplot(1,2,2)
plt.plot(x,y,'g', x, z, 'b', z,a,'r')
plt.xlim([0,5])
plt.ylim([-0.5,0.5])

plt.savefig('examples.png')
plt.show()
```
